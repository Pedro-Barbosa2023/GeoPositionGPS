```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

def load_mnist():
    return tf.keras.datasets.mnist.load_data()

def process_image(image):
    flipped_matrix = np.apply_along_axis(np.flip, 1, image)
    transposed_matrix = np.transpose(flipped_matrix)
    return transposed_matrix

def display_image(matrix, label):
    plt.imshow(matrix, cmap='Blues')
    plt.title(f'Digito: {label}')
    plt.show()

# Carregamento do dataset MNIST
(x_test, y_test), (x_train, y_train) = load_mnist()

# Seleção de uma imagem
selected_image_index = 12
input_matrix = x_train[selected_image_index, :, :]
output_matrix = process_image(input_matrix)

# Exibição da imagem processada
display_image(output_matrix, y_train[selected_image_index])

# Dimensões da imagem
image_rows, image_cols = 28, 28

# Variáveis
batch_size = 64
num_classes = 10
epochs = 5  # Corrigido de 'epoches' para 'epochs'

x_train = x_train.reshape((x_train.shape[0], image_rows, image_cols, 1))
x_test = x_test.reshape((x_test.shape[0], image_rows, image_cols, 1))
i_shape = (image_rows, image_cols, 1)

# Definindo a escala
x_train = (x_train/255.0)
x_test = (x_test/255.0)

y_train = tf.keras.utils.to_categorical(y_train, num_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes)

# Estrutura do modelo CNN
cnn_model = models.Sequential([
    layers.Conv2D(20, kernel_size=(5,5), activation='relu', input_shape=i_shape),
    layers.MaxPooling2D(pool_size=(2,2)),
    layers.Conv2D(40, kernel_size=(3,3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),
    layers.Dropout(0.25),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')
])

cnn_model.summary()

# Compilar o modelo
cnn_model.compile(
    loss=tf.keras.losses.categorical_crossentropy,
    optimizer=tf.keras.optimizers.Adadelta(),  # Corrigido de 'optmizers' para 'optimizers'
    metrics=['accuracy']
)

# Trainar o modelo
cnn_history = cnn_model.fit(
    x_train, y_train,
    batch_size=batch_size,
    epochs=epochs,  # Corrigido de 'epoches' para 'epochs'
    validation_split=0.2
)

# Plot da curva de aprendizado
plt.plot(cnn_history.history['accuracy'], label='Acurácia do treino', color='red')
plt.plot(cnn_history.history['val_accuracy'], label='Acurácia da validação', color='purple')
plt.xlabel('Epocas')
plt.ylabel('Acurácia')
plt.legend()
plt.show()

# Avaliar o modelo
cnn_model.evaluate(x_test, y_test)

# Previsões do modelo
cnn_pred = cnn_model.predict(x_test)
print(cnn_pred[:50])

# Obter pesos do modelo
weights = cnn_model.get_weights()

# Plotar filtros
plt.imshow(weights[0][:, :, 0, 0], cmap='gray')  # filtro 1 da camada 1
plt.show()

plt.imshow(weights[0][:, :, 0, 2], cmap='gray')  # filtro 3 da camada 1
plt.show()
```
